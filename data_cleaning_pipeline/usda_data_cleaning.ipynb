```python
df_unemp = pd.read_csv("Unemployment2023.csv")
df_pov = pd.read_csv("Poverty2023.csv")
```

```python
df_unemp['Attribute'].unique()
```

```python
# removing records where Area names don't have commas, to get county level records
# filters 329726 entries to 324438
df_unemp_filtered = df_unemp[df_unemp['Area_Name'].str.contains(",")]
```

```python
df_unemp_filtered.head()
```

```python
df_unemp_filtered[['Feature', 'Year']] = df_unemp_filtered['Attribute'].str.extract(r'(.+)_(\d{4})')
df_unemp_filtered['Year'] = df_unemp_filtered['Year'].astype(int)
```

```python
df_unemp_filtered['Year'].unique()
```

```python
# split attribute column to extract year and feature variable
df_unemp_filtered[['Feature', 'Year']] = df_unemp_filtered['Attribute'].str.extract(r'(.+)_(\d{4})')
df_unemp_filtered['Year'] = df_unemp_filtered['Year'].astype(int)
df_unemp_filtered
```

```python
# pivot table to get one row per
df_unemp_filtered = df_unemp_filtered.pivot_table(
    index=['FIPS_Code', 'State', 'Area_Name', 'Year'],
    columns='Feature',
    values='Value'
).reset_index()

df_unemp_filtered.head(10)
```

```python
# get the most recent year of population volumes
df_unemp_final = df_unemp_filtered[df_unemp_filtered['Year'] == 2023]
df_unemp_final.head()
```

```python
# the data is incomplete where there are missing records for multiple years where only the year 2023 contains all the columns including median household income
df_unemp_final.info()
```

```python
df_unemp_final = df_unemp_final.drop(
    columns=[
        'Med_HH_Income_Percent_of_State_Total',
        'Median_Household_Income',
        'Urban_Influence_Code'
    ],
    errors='ignore'
)
```

```python
# for the most recent year of data there is not record of median household income
# decided to use the most recent year of data for population volumes (2023) and then the most recent year of data for income (2022)
df_income_2022 = df_unemp_filtered[df_unemp_filtered['Year'] == 2022][
    ['FIPS_Code', 'Median_Household_Income',
     'Med_HH_Income_Percent_of_State_Total']
]

df_unemp_final = df_unemp_final.merge(
    df_income_2022,
    on='FIPS_Code',
    how='left'
)
```

```python
df_unemp_final.info()
```

```python
numeric_cols = [
    'Civilian_labor_force', 'Employed', 'Unemployed',
    'Unemployment_rate', 'Metro', 'Rural_Urban_Continuum_Code',
    'Median_Household_Income', 'Med_HH_Income_Percent_of_State_Total'
]

df_numeric = df_unemp_final[numeric_cols]
```

```python
# will impute the missing numeric values so counties will not be neglected
imputer = KNNImputer(n_neighbors=5)
df_imputed = imputer.fit_transform(df_numeric)

df_unemp_final[numeric_cols] = df_imputed
```

```python
df_unemp_final.info()
```

```python
df_unemp_final.describe()
```

```python
features = [
    'FIPS_Code',
    'State',
    'Area_Name',
    'Year',
    'Civilian_labor_force',
    'Employed',
    'Unemployed',
    'Unemployment_rate',
    'Median_Household_Income',
    'Med_HH_Income_Percent_of_State_Total'
]

df_unemp_final = df_unemp_final[features]
```

```python
df_unemp_final.head(10)
```

```python
df_unemp_final.to_csv("Cleaned_Unemployment2023.csv", index=False)
```
