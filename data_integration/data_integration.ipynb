```python
import pandas as pd
```

```python
#loading all the cleaned data
df_fema = pd.read_csv('cleaned_fema_disasters.csv')
df_poverty = pd.read_csv('Cleaned_Poverty2023.csv')
df_unemployment = pd.read_csv('Cleaned_Unemployment2023.csv')
df_pop = pd.read_csv('cleaned_population_estimates.csv')
df_noaa = pd.read_csv('cleaned_noaa_events.csv')
```

```python
df_fema['incident_year'] = pd.to_datetime(df_fema['incident_start_date']).dt.year
```

```python
#header cleaning for all
for df in [df_fema, df_poverty, df_unemployment, df_pop, df_noaa]: df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
#merging based on fips code
df_master = df_fema.merge(df_poverty, on='fips_code', how='left')
df_master = df_master.merge(df_unemployment, on='fips_code', how='left')
df_master = df_master.merge(df_pop[['fips_code', 'total_population', 'total_population_norm']], on='fips_code', how='left')
```
